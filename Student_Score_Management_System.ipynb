Student Score Management System

Scenario

You are building a small system for a lecturer to manage student exam scores.

The program should:

1. Allow the lecturer to enter student scores.

2. Sort the scores using bubble sort.

3. Use a lambda function to calculate grade status.

4. Use recursion to calculate the class total score.

5. Allow the lecturer to search for a score using:

6. inear search

7. Binary search

8. Use loops and conditions to control the program.





```python
scores = []
# list to store scores
```

```python
#loop to enter student values
while True:
  score = input('Enter the student marks(or q to quit): ')
  if score == 'q':
    break
  if not score.isdigit():
    print("invalid input! Please enter numbers.")
    continue
  score = int(score)
  if score < 0 and score > 100 :
    print("Score must be within 0 and 100")
    pass
  else:
    scores.append(score)
  print("Scores entered: ",scores)
```

```python
#bubble sort to sort the scores
```

```python
def bubble_sort(arr_n):
  n = len(arr_n)

  while True:
    swapped = False
    for i in range(n-1):
      if arr_n[i] > arr_n[i+1]:
        arr_n[i],arr_n[i+1] = arr_n[i+1],arr_n[i]
        swapped = True
    n = n-1
    if swapped == False:
      break
bubble_sort(scores)
```

Lambda functions to calculate grade status

```python
check_pass = lambda x : "Pass" if x >= 50 else "Fail"
print("\nStudents Results:")
for s in scores:
  print(s, "-", check_pass(s))
```

Use recursion to calculate the class total score.


```python
def total_score(arr_n):
  if len(arr_n) == 0:
    return 0
  return arr_n[0] + total_score((scores))

  print("Total class score:", total_score(scores))
```

Allow the teacher to look for a score using linear search:

```python
def linear_search(arr_n,target):
  for i in range(len(arr_n)):
    if arr_n == target:
      return i
    return -1
```

Allow the teacher to look for a score using binary search

```python
def binary_search(arr_n,target):
  low = 0
  high = len(arr_n) - 1

  while low <= high:
    mid = (high + low) // 2
    if arr_n[mid] == target:
      return mid
    elif arr_n[mid] < target:
      low = mid + 1
    else:
      high = mid - 1
  return -1
```

```python
target = int(input("\nEnter score to search: "))

index = linear_search(scores, target)

if index != -1:
    print("Found using Linear Search at position", index)
else:
    print("Score not found using Linear Search")


index = binary_search(scores, target)

if index != -1:
    print("Found using Binary Search at position", index)
else:
    print("Score not found using Binary Search")
```
