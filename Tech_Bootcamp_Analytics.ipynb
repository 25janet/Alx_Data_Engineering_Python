A tech bootcamp records the courses taken by students and their scores.



```python
bootcamp = {
    "Alice": {"Python": 85, "AWS": 78, "Docker": 92},
    "Bob": {"Linux": 88, "AWS": 90, "Terraform": 84},
    "Charlie": {"Python": 95, "Kubernetes": 80, "Docker": 89},
    "Diana": {"AWS": 91, "Terraform": 87, "Python": 82},
    "Eve": {"Docker": 86, "Linux": 79, "Kubernetes": 88}
}
```

Your program should analyze the dataset and produce the following:

* The set of all courses offered in the bootcamp

* The average score for each student

* The student with the highest overall average score

* The course with the highest average score across all students

* A dictionary showing how many students are enrolled in each course

* A set of students who are taking both Python and Docker

* A dictionary mapping each student to their highest scoring course

* Whether every student scored at least 80 in one course

* A sorted list of all courses based on their average score (highest to lowest)

```python
unique_courses = set()
for courses in bootcamp.values():
  for key in courses.keys():
      unique_courses.add(key)
print(unique_courses)
```

The average score for each student.

```python

for name , courses in bootcamp.items():
  sum_course = 0
  count_course = 0
  print(f"Name : {name}")
  for values in courses.values():
    sum_course += values
    count_course += 1
  average_course = sum_course / count_course
  print(f"The average score is: {average_course:.2f}")
```

The student with the highest overall average score

```python
top_student = None
max_avg =  0
for name,courses in bootcamp.items():
  sum_course = 0
  count_course =  0

  for values in courses.values():
    sum_course += values
    count_course += 1
  average_course = sum_course / count_course
  if average_course > max_avg :
    max_avg = average_course
    top_student = name
print(f"{top_student} has the highest average score, which is {max_avg}")
```

The course with the highest average score across all students


```python
course_totals = {}
course_counts = {}

for student_scores in bootcamp.values():
    for course, score in student_scores.items():
        course_totals[course] = course_totals.get(course, 0) + score
        course_counts[course] = course_counts.get(course, 0) + 1


course_averages = {}
for course in course_totals:
    course_averages[course] = course_totals[course] / course_counts[course]


highest_avg_course = max(course_averages, key=course_averages.get)
highest_avg_score = course_averages[highest_avg_course]

print(f"Course with highest average: {highest_avg_course} ({highest_avg_score:.2f})")
```

A dictionary showing how many students are enrolled in each course

```python
course_enrollment = {}
for student_courses in bootcamp.values():
  for course in student_courses.keys():
    course_enrollment[course] = course_enrollment.get(course,0) + 1
print(f"Number of students enrolled per course: {course_enrollment}")
```

A set of students who are taking both Python and Docker

```python
student_python_docker = set()
for student ,courses in bootcamp.items():
  if 'Python' in courses and 'Docker' in courses :
    student_python_docker.add(student)
print("Students taking both Python and Docker:", student_python_docker)
```

A dictionary mapping each student to their highest scoring course.

```python
for student,courses in bootcamp.items():
  max_course_score = 0
  top_course_name = None
  print(f"Name: {student}")
  for course_name , course_score in courses.items():
    if course_score > max_course_score:
      max_course_score = course_score
      top_course_name = course_name
  print(f"The highest course {top_course_name} with a score of {max_course_score}")
```

Whether every student scored at least 80 in one course

```python
all_courses_above_80 = True
for student in bootcamp.values():
  if 80 not in student.values():
    all_courses_above_80 = False
    break
print(all_courses_above_80)
```
