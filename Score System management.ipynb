Exercise: Student Result Analyzer (with Data Structures)
Scenario

A lecturer wants a program that:

1. Stores student names and scores

2. Sorts scores using bubble sort

3. Uses lambda to assign grades

4. Uses recursion to compute total marks

5. Uses linear search to find a student

6. Uses binary search to find a score

7. Uses lists, dictionaries, tuples, and sets

```python
students = {}
scores = []
grade_range = (50,100)
```

```python
#input for Student name and scores

while True:
  name = input('Enter the student name(or q to quit): ')
  if name == 'q':
    break
  score = input('Enter the student score(or q to quit): ')
  if score == 'q':
    break
  score = int(score)
  if score < 0 and score > 100 :
    print("Invalid input!Score must range within 0 and 100")
    continue
  students[name] = score
  scores.append(score)

print(f'Students Dictionary: {students}')
```

```python
#set of unique scores
unique_scores = set(scores)
print(f"The set of unique scores: {unique_scores}")
```

```python
# sort scores using bubble sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        # If no elements were swapped, the list is already sorted
        if not swapped:
            break


bubble_sort(scores)
print("Sorted scores:", scores)
```

```python
#lambda function to check whether they have fail or pass
check_pass = lambda x: "Pass" if grade_range[0] <= x <= grade_range[1] else "Fail"
print("\nResults:")
for name, score in students.items():
  print(name, score, check_pass(score))
```

```python
#recursion to comute total marks
def total_marks(arr):
  if len(arr) == 0:
    return 0
  return arr[0] + total_marks(arr[1:])

print("\nTotal class marks:", total_marks(scores))
```

```python
#Linear search to find a student
def linear_search(arr,target):
  for name in arr:
    if name == target:
      return arr[name]
  return None


student_to_find = input("\nStudent to search for: ")
result = linear_search(students,student_to_find)

if result is not None:
  print("Score:",result)
else:
  print("Student not found.")

```

```python
def binary_search(arr,target):
  low = 0
  high = len(arr) -1
  while low <= high:
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid

    elif arr[mid] < target:
        low = mid + 1

    else:
        high = mid - 1

    return -1
target_score = int(input("\nEnter score to search: "))

position = binary_search(scores, target_score)

if position != -1:
    print("Score found at position", position)
else:
    print("Score not found")
```
