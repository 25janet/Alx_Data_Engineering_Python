A tech training center records the technologies that each student knows.

Some students may list the same skill more than once,
  
  but the system must treat skills as unique.

Your program should determine:

The complete set of unique technologies known across all students.

* The technologies shared by Alice and Charlie.

* The technologies that Bob knows but Alice does not.

* The technologies known by exactly one student in the entire dataset.

* Whether every student knows at least one cloud technology (AWS or Terraform).

* The student who knows the highest number of unique technologies.

 * A set of students who know Docker but do not know Kubernetes.

* A sorted list of all technologies offered by the training center.

* A mapping showing each student and their unique skills count.





```python
students = {
    "Alice": ["Python", "AWS", "Docker", "Python"],
    "Bob": ["Linux", "AWS", "Terraform", "Linux"],
    "Charlie": ["Python", "Kubernetes", "Docker"],
    "Diana": ["AWS", "Terraform", "Python"],
    "Eve": ["Docker", "Linux", "Kubernetes", "Docker"]
}
```

Complete set of  *unique* technologies

```python
unique_technologies = set()
for skill_list in students.values():
  for skill in skill_list:
    unique_technologies.add(skill)
print(f"All technologies: {sorted(unique_technologies)}")
print(f"The total account of technology provided is {len(unique_technologies)}")
```

The technology shared by Alice and Charlie.

```python
alice_set = set(students['Alice'])
charlie_set = set(students['Charlie'])
shared_set = alice_set.intersection(charlie_set)
print(f"Alice and Charlie shared skill is {shared_set}")
```

The technolgies Bob knows and Alice does not.

```python
Bob_set = set(students['Bob'])
Alice_set = set(students["Alice"])
Bob_unique_set = Bob_set.difference(Alice_set)
print(f"Bob uniques set vs Alice: {Bob_unique_set}")
```

The technologies known by exactly one student in the entire dataset.

```python
technology_count = {}
for student in students:
  skills = set(students[student])
  for tech in skills:
    if tech not in technology_count:
      technology_count[tech] = 1
    else:
      technology_count[tech] += 1
unique_tech = set()
for skill in technology_count:
  if technology_count[tech] == 1:
    unique_tech.add(tech)
print("Technologies known by exactly one student:", unique_tech)

```

Whether every student knows at least one cloud technology (AWS or Terraform).

```python
cloud_tech = {"AWS", "Terrafoam"}
all_have_cloud = True

for student in students:
  skills = set(students[student])
  if skills.isdisjoint(cloud_tech):
    all_have_cloud = False
    break
print("Every student knows a cloud technology:", all_have_cloud)
```

The student who knows the highest number of unique technologies.


```python
top_student = None
max_skills = 0
for student in students:
  unique_skills = set(students[student])
  skill_count = len(unique_skills)
  if skill_count > max_skills:
    max_skills = skill_count
    top_student = student
print("Top skilled student:", top_student)
print("Number of unique technologies:", max_skills)
```

A set of students who know Docker but do not know Kubernetes.

```python
docker_only_set = set()
for student in students:
  skills = set(students[student])
  if 'Docker' in skills and 'Kubernetes' not in skills:
    docker_only_set.add(student)
print("Students who know Docker but not Kubernetes:", docker_only_set)
```

A sorted list of all technologies offered by the training center.

```python
unique_tech = set()
for skill_list in students.values():
  for skill in skill_list:
    unique_tech.add(skill)
print(unique_tech)
sorted_unique_tech = sorted(unique_tech)
print(sorted_unique_tech)
```

A mapping showing each student and their unique skills count.

```python
for student in students:
  name = student
  unique_skills = set(students[student])
  skill_count = len(unique_skills)
  print(f"{name} has {skill_count} unique courses")
```
