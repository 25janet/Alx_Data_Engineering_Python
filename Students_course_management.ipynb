A university stores students and their courses in a nested list.

```python
students = [
    ["Alice", ["Python", "AWS", "SQL"]],
    ["Bob", ["Linux", "Networking"]],
    ["Charlie", ["Python", "Docker", "Kubernetes"]],
    ["Diana", ["SQL"]]
]
```

```python
for student in students:
  name = student[0]
  Course = student[1][0:]
  print(f"Students:{name}")
  print(f"Course:{Course}")
```

Count total courses per student


```python
for student in students:
  name = student[0]
  course = student[1][0:]
  print(f"{name} is enrolled in {len(course)} courses")
```

Find students taking Python

```python
for student in students:
  name = student[0]
  course = student[1][0:]
  if 'Python' in course:
    print(f"{name} is taking Python.")
```

Charlie decides to drop Docker.


```python
for student in students:
  name = student[0]
  course = student[1]
  if name == "Charlie":
    if "Docker" in course:
      course.remove("Docker")
    print(name)
    print(course)
```

```python
top_students = None
max_courses = -1
for student in students:
  name = student[0]
  courses = student[1][0:]
  num_courses = len(courses)
  if num_courses >  max_courses:
    max_courses = num_courses
    top_students = name
if top_students:
  print(f"{top_students} has the most courses: {max_courses}")
```

```python
unique_courses = []
for student in students:
  for course in student[1]:
    if course not in unique_courses:
      unique_courses.append(course)
print(unique_courses)
```
